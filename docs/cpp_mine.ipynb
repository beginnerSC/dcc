# C++ Notes

## Overloading Operator `->`

* The `->` operator should return a pointer, as the compiler translates `it->something` to `(it.operator->())->something`

## Overloading by Return Type?

* overload 只能 by arguemnt type，不能 by return type
* 但下面這樣 overload 是可以的，因為 const 會改變 this （implicit argument）的 type，所以 signature 不一樣
* index const Vector 時會呼叫 const version，index non-const Vector 時會呼叫 non-const version

```cpp
  int& operator[](size_t i);
  const int& operator[](size_t i) const;
```
